```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pickle

save = False
idx = pd.IndexSlice

# Get result table
compdf = pickle.load(open('hist_performance.p', 'rb'))

# table report
table = pd.read_pickle('./result/final_table.p')
table.loc[:,idx[['btc', 'ensemble', 'switch', 'var_evt', 'var_norm', 'varspread'], 'total_ret']]
index_model = list(table.columns.levels)[0]
quantiles = list(table.index.levels)[0]
window = list(table.index.levels)[1]

# change index
newtable = pd.DataFrame(index =pd.MultiIndex.from_product([quantiles, window, index_model]), columns = list(table.columns.levels)[1])
qws = table.index
for qw in qws:
    for m in index_model:
        newtable.loc[(qw[0], qw[1], m)] = table.loc[qw, idx[m,:]][m]

to_save = newtable.loc[idx[:,:,['btc', 'ensemble', 'var_norm', 'var_evt', 'varspread', 'switch']], ['total_ret', 'sr', 'sortino', 'mdd', 'VaR', 'vol']]
to_save['total_ret'] = (to_save['total_ret'] + 1) ** (24*365/len(compdf)) - 1
to_save.loc[:, ['total_ret', 'mdd', 'VaR', 'vol']] = to_save[['total_ret', 'mdd', 'VaR', 'vol']] * 100
to_save = to_save.astype('float')
to_save.loc[:, ['VaR', 'vol']] = to_save.loc[:, ['VaR', 'vol']].round(2)
to_save.loc[:, ['sr', 'sortino']] = to_save.loc[:, ['sr', 'sortino']].round(2)
to_save.loc[:, ['total_ret', 'mdd']] = to_save.loc[:, ['total_ret', 'mdd']].round(0).astype(int)
to_save = to_save.loc[idx[:,:,['ensemble', 'var_norm', 'var_evt', 'varspread', 'switch']],:]
if save:
    to_save.to_csv('final_back_strat_perf.csv')
```
