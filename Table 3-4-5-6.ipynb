```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from result.utils import qs_from_name, get_col_from_wq, get_boxdata
from result.compare import get_table_report
import pickle

save = False
legend = False

# constants
idx = pd.IndexSlice
classifiers = ['mlp', 'lstm', 'garch', 'evtgarch', 'lpa', 'carl', 'ensemble']
order_cols = ['btc'] + classifiers + ['var_norm', 'var_evt', 'varspread']
ws_name = ['woneday', 'w4months', 'w6months']
qs_name = ['01', '025', '05', '10']


# Get result table
compdf = pickle.load(open('hist_performance.p', 'rb'))
```

# Table report

```python
table = get_table_report(compdf, qs_name, ws_name, varspread=True, carl=True, lpa=True, 
                         ensemble=True, var_norm = True, var_evt = True, mlp=True)
if save:
    table = pickle.dump(table, open('test_table.p', 'wb'))
```

## Table 3: Risk-adjusted AUC on the test set

```python
(np.round(table[classifiers].loc[:, idx[:, 'risk_auc']]*100)).astype(int)#.idxmax(1)
```

# cross validation result

```python
# load cv dates
cvdates = pickle.load(open('./result/cvdates.p', 'rb'))
cvtable = {}
for k in cvdates.keys():
    print(k)
    cvtable[k] = get_table_report(compdf.loc[cvdates[k]].dropna(), qs_name, ws_name, varspread=True, carl=True, lpa=True, 
                             ensemble=True, var_norm = True, var_evt = True, mlp=True)
    new_ind = pd.MultiIndex.from_product([[0.01, 0.025, 0.05, 0.1], [24, 2880, 4320]])
    cvtable[k].index = new_ind
if save:
    pickle.dump(cvtable, open('cv_table.p', 'wb'))
```

## Table 4: Average risk-adjusted AUC and variance risk-adjusted AUC over the testing periods

```python
metric = 'risk_auc'
cvperf = pd.DataFrame(columns = pd.MultiIndex.from_product([classifiers, ['mean', 'var']]), 
                      index = table.index)
for qw in table.index:
    boxdata = get_boxdata(cvtable, metric, classifiers, qw)
    arcvperf = [[np.mean(b), np.var(b)] for b in boxdata]
    for i,m in enumerate(classifiers):
        cvperf.loc[qw][m] = arcvperf[i]
(cvperf.astype(float) * 100).round(1)
```

## Table 5: Exceedances of tail risk protection strategies in %

```python
stat = (table[order_cols].loc[:, idx[:, 'exceedance']] * 100).astype(float).round(1).droplevel(1,1)
if save:
    stat.to_csv('exceedance.csv')
stat
```

```python
stat = table[order_cols].loc[:, idx[:,'ret']].droplevel(1,1) * 100
for c in stat.columns:
    stat[c] = stat[c].map('{:.1e}'.format)
if save:
    stat.to_csv('avg_return.csv')
stat
```
